```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt  # Import matplotlib for plotting
from sklearn.metrics.pairwise import cosine_similarity

# Load data
def load_data(train_file, label_train_file, validate_file, Label_validate_file):
    try:
        train_data = pd.read_csv(train_file, header=None).values
        train_labels = pd.read_csv(label_train_file, header=None).values.ravel()
        validate_data = pd.read_csv(validate_file, header=None).values
        validate_labels = pd.read_csv(Label_validate_file, header=None).values.ravel()

        # Ensure validation labels have the same number of classes as training labels
        num_classes = len(np.unique(train_labels))
        validate_labels = pd.get_dummies(validate_labels)
        if len(validate_labels.columns) < num_classes:
            for i in range(num_classes - len(validate_labels.columns)):
                validate_labels[f'Class_{i}'] = 0
        elif len(validate_labels.columns) > num_classes:
            validate_labels = validate_labels.iloc[:, :num_classes]

        return train_data, train_labels, validate_data, validate_labels.values
    except Exception as e:
        print("Error loading data:", e)
        return None, None, None, None

# Initialize population
def initialize_population(population_size, chromosome_length):
    return np.random.randint(2, size=(population_size, chromosome_length))

# Evaluate fitness
def evaluate_fitness(population, train_data, train_labels, validate_data, validate_labels):
    fitness_scores = []
    for chromosome in population:
        selected_indices = np.where(chromosome == 1)[0]  # Get indices of selected features
        selected_features = train_data[:, selected_indices]
        selected_validation_features = validate_data[:, selected_indices]

        # Check if any selected features or validation features are present
        if selected_features.shape[1] == 0 or selected_validation_features.shape[1] == 0:
            fitness_scores.append(0)
            continue

        # Calculate cosine similarity using selected features
        cos_similarities = cosine_similarity(selected_features, selected_validation_features)

        # Compute accuracy
        predicted_labels = np.argmax(cos_similarities, axis=0)
        accuracy = np.mean(predicted_labels == validate_labels.argmax(axis=1))
        fitness_scores.append(accuracy)

    return np.array(fitness_scores)

# Random selection
def random_selection(population, fitness_scores):
    selected_index = np.random.choice(len(population), p=fitness_scores/np.sum(fitness_scores))
    return population[selected_index]

# Mutation
def mutation(child, mutation_rate):
    for i in range(len(child)):
        if np.random.rand() < mutation_rate:
            child[i] = 1 - child[i]  # Flip 0 to 1 or 1 to 0
    return child

# Reproduction
def reproduce(x, y):
    n = len(x)
    c = np.random.randint(1, n)  # Randomly choose a crossover point
    child = np.concatenate((x[:c], y[c:]))  # Concatenate substrings from parents
    return child

# Genetic algorithm
def genetic_algorithm(train_file, label_train_file, validate_file, Label_validate_file,
                      num_features, population_size, max_generations, mutation_rate):
    train_data, train_labels, validate_data, validate_labels = load_data(train_file, label_train_file,
                                                                         validate_file, Label_validate_file)
    if train_data is None or train_labels is None or validate_data is None or validate_labels is None:
        print("Error: Failed to load data.")
        return

    num_classes_train = len(np.unique(train_labels))
    num_classes_validate = validate_labels.shape[1]
    print("Number of classes in training labels:", num_classes_train)
    print("Number of classes in validation labels:", num_classes_validate)
    if num_classes_train != num_classes_validate:
        print("Error: Number of classes in training and validation labels do not match.")
        return

    population = initialize_population(population_size, num_features)
    fitness_history = []  # List to store fitness scores over generations

    for generation in range(max_generations):
        fitness_scores = evaluate_fitness(population, train_data, train_labels, validate_data, validate_labels)
        fitness_history.append(np.mean(fitness_scores))

        new_population = []
        for _ in range(population_size):
            x = random_selection(population, fitness_scores)
            y = random_selection(population, fitness_scores)
            child = reproduce(x, y)
            if np.random.rand() < mutation_rate:
                child = mutation(child, mutation_rate)
            new_population.append(child)

        population = np.array(new_population)

    best_chromosome = population[np.argmax(fitness_scores)]
    final_fitness = evaluate_fitness(np.array([best_chromosome]), train_data, train_labels, validate_data, validate_labels)[0]
    qualified_features = np.where(best_chromosome == 1)[0]
    print(f"\nFinal optimized feature vector accuracy: {final_fitness*100:.2f}%")
    print("Number of qualified feature vectors:", len(qualified_features))
    print("Qualified feature vectors:", qualified_features)

    # Plot fitness history
    plt.plot(fitness_history)
    plt.title('Fitness Score over Generations')
    plt.xlabel('Generation')
    plt.ylabel('Fitness Score')
    plt.show()

# Example usage
genetic_algorithm("train.csv", "label_train.csv", "validate.csv", "Label_validate.csv",
                  num_features=1433, population_size=65, max_generations=100, mutation_rate=0.01)
```
